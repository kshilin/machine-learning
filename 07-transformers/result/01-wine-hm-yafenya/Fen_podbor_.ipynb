### Область работы 2 (поиск  модели .... )

```python
# Данный блок(и) НЕ выполняются преподавателем в области 2
# блок(и) предназначены для поиска лучшей модели 
# должен быть понятен и очевиден отбор параметров модели
# оставляйте свои комментарии и разъяснения
# 
# Запрещается размещать данные блоки за пределами обасти 2
# Все блоки данной области должны быть выполнены
#
# ЗАПРЕЩАЕТСЯ ИСПОЛЬЗОВАТЬ ТЕСТОВЫЙ НАБОР
#
# Путь к тренировочному набору
# 

path_train = pd.read_csv('train.csv') # содержит только имя файла, без имен папок !!!
display(path_train)
```

### Я посчитал, что нужная метрика - это точность. Наиболее рационально определить, сколько плохих вин мы назвали хорошими, чем сколько хороших вин обозначили как плохие

```python
wine_train = path_train
wine_train.describe()
```

```python
wine_train = wine_train.drop_duplicates()
wine_train.describe()
```

### После долгих попыток с перебором вариантов нужных нам столбцов, пришел к выводу, что из-за столбца density фонит результат, поэтому убираем его

```python
wine_train["quality"] = np.where(wine_train['quality'] >= 7, 1, 0)
wine_train.drop('density', inplace=True, axis= 1 )
wine_train.head()
```

```python
x = wine_train.drop('quality',axis = 1)
y = wine_train['quality']
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42, stratify = y)
```

 ##### Попытки получить что-то интересное с помощью метода КНН с горечью врезались в значение точности = 0.5-0.6, поэтому данные попытки были оставлены, а на помощь пришла логистическая регрессия, которая сразу же показала результат в 0.73. Далее методом проб и ошибок, подбором лучших параметров, получилось следующее: 

```python
pipe = Pipeline([('preprocessing', StandardScaler()), 
                 ('clf', LogisticRegression())])

scaling = [MinMaxScaler(), StandardScaler(), RobustScaler(), Normalizer()]

param_grid =[
    {'preprocessing': scaling,'clf__penalty': ['l2'], 
         'clf__solver': ['newton-cg' ,'lbfgs', 'liblinear', 'sag', 'saga'], 
     'clf__C':[0.001,0.1,0.5,1], 'clf__max_iter': [10,100,1000]},
    {'preprocessing': scaling,'clf__penalty': ['l1'], 
         'clf__solver': ['liblinear'], 'clf__C':[0.001,0.1,0.5,1], 'clf__max_iter': [10,100,1000]},
    {'preprocessing': scaling,'clf__penalty': ['none'], 
         'clf__solver': ['lbfgs','newton-cg'],'clf__C':[0.001,0.1,0.5,1], 'clf__max_iter': [10,100,1000]}
    ]

grid = GridSearchCV(pipe, param_grid, cv = 5, return_train_score = True, scoring = 'precision')
grid.fit(x_train, y_train)
grid
```

### Найдем лучший набор параметров с помощью решетки

```python
grid_result = pd.DataFrame(grid.cv_results_).sort_values("rank_test_score").T
grid_result
```

### Вставим данные параметры в нашу модель

```python
pipe_new = make_pipeline(StandardScaler(), LogisticRegression(C = 0.001, max_iter = 100, 
                                                                   penalty = 'l2', solver = 'saga'))
pipe_new.fit(x_train,y_train)
y_new = pipe_new.predict(x_test)
print('Правильность на тестовом наборе:',  precision_score(y_test, y_new))
```

### Область работы 2 (выполнение лучшей модели)

```python
# Данный блок(и) в области 2 выполняется преподавателем
#
# В области находится одна, единственная, итоговая модель машинного обучения с однозначными, 
# зафиксированными параметрами
#
# В данной области категорически запрещается искать, выбирать, улучшать, оптимизировать, 
# тюниговать и т.д. модель машинного обучения
```

```python
# Путь к тренировочному набору
path_train = 'train.csv' # содержит только имя файла, без имен папок
# Путь к тестовому набору
path_test = 'test.csv' # содержит только имя файла, без имен папок
```

```python
# Блок(и) обучения и поверки модели
```

```python
df_train = pd.read_csv(path_train)
df_train['quality'] = df_train['quality'].apply(lambda x: 0 if x < 7 else 1)
df_train.head()
```

```python
df_test = pd.read_csv(path_test)
df_test.head()
```

```python
pipe_end = make_pipeline(RobustScaler(), KNeighborsClassifier(leaf_size=5, n_jobs=-1, n_neighbors=98, p=2, weights='uniform', algorithm='auto'))
X = df_train.drop(columns = ['quality'])
y = df_train['quality']
display(X.head())
display(y.head())
pipe_end.fit(X,y)
```

```python
# Блок предсказания с использованием тестового набора
```

```python
dvumern = pipe_end.predict_proba(df_test)
y_final_predict = (dvumern[:, 1] > 0.714).astype(int)
print(y_final_predict)
```

```python
# Название вектора предсказанных значений  y_predict полученого на основании тестового набора
y_predict = y_final_predict
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score


y_true = np.where(pd.read_csv('test_Y_true.csv').values>=7,1,0)
matrix=confusion_matrix(y_true, y_predict)
ConfusionMatrixDisplay(matrix).plot();
precision_score(y_true, y_predict)
```

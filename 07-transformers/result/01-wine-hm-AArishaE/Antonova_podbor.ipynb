```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, Normalizer
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, KFold, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB 
from sklearn.metrics import classification_report
```

### Область работы 2 (поиск  модели .... )

```python
# Данный блок(и) НЕ выполняются преподавателем в области 2
# блок(и) предназначены для поиска лучшей модели 
# должен быть понятен и очевиден отбор параметров модели
# оставляйте свои комментарии и разъяснения
# 
# Запрещается размещать данные блоки за пределами обасти 2
# Все блоки данной области должны быть выполнены
#
# ЗАПРЕЩАЕТСЯ ИСПОЛЬЗОВАТЬ ТЕСТОВЫЙ НАБОР
#
# Путь к тренировочному набору
# 

path_train = ... # содержит только имя файла, без имен папок !!!
```

```python
path_train = 'train.csv'
df = pd.read_csv(path_train)

df.info()
df.describe()
```

```python
df.tail(10)
```

```python
df['quality']= np.where(df['quality'] < 7,0,1)
df.tail(10)
```

```python
df.isnull().sum() #Проверка данных на ошибки и пропущенные значения
```

```python
X = df.drop(['quality'],axis=1).values
y = df['quality'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # мы разделяем наши данные на обучающий и тестовый наборы

knn = KNeighborsClassifier() 
knn.fit(X_train,y_train)  #обучаем эту модель на обучающем наборе данных с использованием метода фит
scrtrain = knn.score(X_train , y_train) #позволяет нам увидеть, насколько хорошо модель работает на данных
scrtest = knn.score(X_test, y_test) #и как она предсказывает на новых, ранее не виденны

print("тренировочный набор: {:.3f}".format(scrtrain))
print("правильность на тестовом наборе: {:.3f}".format(scrtest))

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
df['quality'].value_counts(normalize = True) #как будто мы наблюдаем дисбаланс классов.....
#Класс 0 (качество менее 7) 64%.
#Класс 1 (качество 7 и выше)36%.
```

#### Используем  precision, тк она показывает, какой процент вин, которые модель предсказала как хорошие, действительно оказались хорошими, те мы хотим избежать покупки вина низкого качества и купить только хорошее
Те хотим не тратить деньги на плохое вино, поэтому лучше удостовериться, что когда модель говорит, что вино хорошее, она права. 
наша цель купить вино хорошего качества (7 и более)

```python
# KNN
pipe = Pipeline([
    ('preprocessing', StandardScaler()), #создаем конвейер, обработываем данные с помощью метода StandardScaler(), который стандартизирует функции, делая среднее значение равным 0 и стандартное отклонение равным 1
    ('clf', KNeighborsClassifier())
])

kfold = StratifiedKFold(n_splits=10, shuffle=True)#stratifiedKFold для разделения набора данных на 10 блоков, сохраняя пропорции классов в каждом блоке

param_grid = [
    {
        'preprocessing': [MinMaxScaler(), StandardScaler(),RobustScaler(),Normalizer()], #проверяем разные методы масштабирования 
        'clf__n_neighbors': range(3, 51, 7), #соседи с шагом 7 , от 3 до 51
        'clf__weights': ['uniform', 'distance'], #uniform: все соседи вносят равный вклад в предсказание
#distance: веса обратно пропорциональны расстоянию. То есть объекты, которые находятся ближе к интересующей точке, будут оказывать на неё большее влияние, чем те, которые дальше

        'clf__p': [1,2,3]
    }
]

grid2 = GridSearchCV(pipe, param_grid, cv=kfold, scoring='precision', return_train_score=True)
grid2.fit(X_train, y_train)

print("----------------- Обучили и тестировали -------------------")
print("Наилучшие параметры:\n{}\n".format(grid2.best_params_))
print("Средняя правильность для наилучшей модели кроссвалидации на валидационных тестовых наборах: {:.6f}\n".format(grid2.best_score_))
print("Правильность для наилучшей модели на тестовом наборе: {:.6f}\n".format(grid2.score(X_test, y_test)))
grid2results = pd.DataFrame(grid2.cv_results_)
display(grid2results.sort_values(["rank_test_score"]).T)
```

```python
y_pred_knn = grid2.predict(X_test)
print("Метрики для KNN:")
print(classification_report(y_test, y_pred_knn))
```

```python
# логрег 
pipe = Pipeline([
    ('preprocessing', MinMaxScaler()), 
    ('clf', LogisticRegression(class_weight='balanced'))
])
kfold = KFold(n_splits=10, shuffle=True)
scaling = [ MinMaxScaler(), StandardScaler(),RobustScaler(), Normalizer()]


param_grid = param_grid = [
    {'preprocessing': scaling, 'clf__penalty': ['l2'], 
     'clf__solver': ['newton-cg' ,'lbfgs', 'liblinear', 'sag', 'saga'], 
     'clf__max_iter': [10000]},  #контролирует максимальное количество итераций, иначе вылезает предпреждение 
    {'preprocessing': scaling, 'clf__penalty': ['l1'], 
     'clf__solver': ['liblinear'], 
     'clf__max_iter': [10000]},  # 
    {'preprocessing': scaling, 'clf__penalty': [None], 
     'clf__solver': ['lbfgs','newton-cg'], 
     'clf__max_iter': [10000]},  # 
]

grid1 = GridSearchCV(pipe, param_grid, cv=kfold, scoring='precision',return_train_score=True)
grid1.fit(X_train, y_train)

print("----------------- Обучили и тестировали -------------------")
print("Наилучшие параметры:\n{}\n".format(grid1.best_params_))
print("Средняя правильность для наилучшей модели кроссвалидации на валидационных тестовых наборах: {:.6f}\n".format(grid1.best_score_))
print("Правильность для наилучшей модели на тестовом наборе: {:.6f}\n".format(grid1.score(X_test, y_test)))
grid1results = pd.DataFrame(grid1.cv_results_)
display(grid1results.sort_values(["rank_test_score"]).T)
```

```python
y_pred_logreg = grid1.predict(X_test)
print("Метрики для логистической регрессии:")
print(classification_report(y_test, y_pred_logreg))
```

```python
# наивный байес
pipe = Pipeline([
    ('preprocessing', MinMaxScaler()), 
    ('clf', GaussianNB())
])

kfold = KFold(n_splits=10, shuffle=True)
scaling = [MinMaxScaler(), StandardScaler(), RobustScaler(), Normalizer()]

# для байеса нет гиперпараметров для регулировки, поэтому сетка состоит только из методов масштабирования
param_grid = [{'preprocessing': scaling}]

grid = GridSearchCV(pipe, param_grid, cv=kfold, scoring='precision', return_train_score=True)
grid.fit(X_train, y_train)

print("----------------- Обучили и тестировали -------------------")
print("Наилучшие параметры:\n{}\n".format(grid.best_params_))
print("Средняя правильность для наилучшей модели кроссвалидации на валидационных тестовых наборах: {:.6f}\n".format(grid.best_score_))
print("Правильность для наилучшей модели на тестовом наборе: {:.6f}\n".format(grid.score(X_test, y_test)))
gridresults = pd.DataFrame(grid.cv_results_)
display(gridresults.sort_values(["rank_test_score"]).T)
```

сравним и ранжируем три разные модели по их точности на тестовом наборе данных

```python
# Для KNN
knn_acc = grid2.score(X_test, y_test)

# Для логистической регрессии
lr_acc = grid1.score(X_test, y_test)

# Для наивного Байеса
nb_acc = grid.score(X_test, y_test)

models = pd.DataFrame({
    'Model' : ['Logistic Regression', 'KNN', 'Naive Bayes'],
    'Score' : [lr_acc, knn_acc, nb_acc]
})

sorted_models = models.sort_values(by='Score', ascending=False)
display(sorted_models)
```

Видим, что KNN выше скор, поэтому выбираю эту модель

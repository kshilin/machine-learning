```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split 
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC

from sklearn import svm
```

```python
X = np.c_[(.4, -.7),
          (-1.5, -1),
          (-1.4, -.9),
          (-1.3, -1.2),
          (-1.1, -.2),
          (-1.2, -.4),
          (-.5, 1.2),
          (-1.5, 2.1),
          (1, 1),
          # --
          (1.3, .8),
          (1.2, .5),
          (.2, -2),
          (.5, -2.4),
          (.2, -2.3),
          (0, -2.7),
          (1.3, 2.1)].T
Y = [0] * 8 + [1] * 8 # это что тут такое?
```

```python
fig, ax = plt.subplots(3, 3, figsize=(10, 5), sharex=True, sharey=True);
kernel = 'rbf'
C = [ 0.1, 1, 10]
gammas = [ 0.1, 1, 10]

for i in range(len(gammas)):
    for j in range(len(C)):
        clf = svm.SVC(kernel=kernel, gamma = gammas[i], C = C[j])
        clf.fit(X, Y)
    
        ax[i, j].scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80,
                    facecolors='none', zorder=10, edgecolors='r')
        ax[i, j].scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired,
                    edgecolors='k')
    
        ax[i,j].axis('tight')
        x_min = -3
        x_max = 3
        y_min = -3
        y_max = 3
    
        XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
        Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])
    
        # Put the result into a color plot
        Z = Z.reshape(XX.shape)
    
        ax[i,j].pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired, shading='auto')
        ax[i,j].contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                    levels=[-.5, 0, .5])
    
        ax[i,j].set_xlim(x_min, x_max)     
        ax[i,j].set_ylim(y_min, y_max)
        ax[i,j].set_xlabel('C = {}'.format(C[j]))
        ax[i,j].set_ylabel('gamma = {}'.format(gammas[i]))
        
        ax[i,j].set_xticks(())
        ax[i,j].set_yticks(())

        fig.suptitle('kernel = {}'.format(kernel))

plt.show()
```
